```python


path = Path(path_meta)
im_path = list(path.rglob('*.jpg'))+list(path.rglob('*.jpeg'))+list(path.rglob('*.png'))
im_names = [p.name for p in im_path]
im_dir = [str(p.parents[0]) for p in im_path]

df = pd.DataFrame(columns=['dir_name','image_name','augmentation','soft_max'])
df['image_name'] = im_names
df['dir_name'] = im_dir
```

```python
newdf = pd.DataFrame(np.repeat(df.values, 4, axis=0))
newdf.columns = df.columns
df=newdf
df.loc[::4,'augmentation'] = 'orig'
df.loc[1::4,'augmentation'] = 'QF90'
df.loc[2::4,'augmentation'] = 'QF65'
df.loc[3::4,'augmentation'] = 'QF40'
# display(df)
df.head(50)

```

```python
for index, row in df.iterrows():
    im_name= str(row['dir_name'])+'/'+(row['image_name'])
    #print(im_name)
    # open the image
    img = Image.open(im_name)
    if(row['augmentation']=='orig'):
        print("use original image")
        # use img for the detector and extract the softmaxes
    elif(row['augmentation']=='QF90'):
        print("QF90")
        img.save(tmp_dir+(row['image_name']), quality=90)
        img2 = Image.open(tmp_dir+(row['image_name'])) # use img2 for the detector and extract the softmaxes
    elif(row['augmentation']=='QF65'):
        print("QF65")
        img.save(tmp_dir+(row['image_name']), quality=65)
        img2 = Image.open(tmp_dir+(row['image_name'])) # use img2 for the detector and extract the softmaxes
    elif(row['augmentation']=='QF40'):
        print("QF40")
        img.save(tmp_dir+(row['image_name']), quality=40)
        img2 = Image.open(tmp_dir+(row['image_name'])) # use img2 for the detector and extract the softmaxes


    # for the example we draw 10 random soft_max (10 classes)
    rand_sm = np.random.rand(10)
    rand_sm = rand_sm/(np.sum(rand_sm))
    row['soft_max']=rand_sm


    
```
